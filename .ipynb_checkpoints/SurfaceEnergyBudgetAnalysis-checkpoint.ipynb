# The Land Surface Energy Budget

This jupyter notebook grabs model output from the Community Earth System Model (CESM: https://www.cesm.ucar.edu) for five locations around the globe. This data has already been pre-processed into a time series of annual mean values for each location, and can be found in the `data/location_files` sub-folder. Raw model data can be downloaded from the Earth System Grid Federation (e.g. at https://esgf-node.llnl.gov/projects/esgf-llnl/). 


In particular, the data here is from the `1pctco2` simulation from the Coupled Model Intercomparison Project 6 (CMIP6: https://www.wcrp-climate.org/wgcm-cmip/wgcm-cmip6). Variables are named following the CMIP6 naming convention (described below). Only variables needed for the surface energy budget are loaded here. 

This script should be run using the python environment described in the `environment.yml` file. 

### Load the packages we need for analysis:

```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import mpl_toolkits
import glob
import pandas as pd
import xarray as xr 
import cartopy
from cartopy.util import add_cyclic_point
import cartopy.crs as ccrs
import scipy.stats as stats 
import scipy as sp
from scipy import signal
import os

from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import matplotlib.ticker as mticker
# from cartopy.mpl.ticker import (LongitudeFormatter, LatitudeFormatter,
#                                 LatitudeLocator, LongitudeLocator)
# import imageio
```

### List the names of the locations we want the data from, and load the datasets into a dictionary that can be accessed by each location name.
(these correspond to file names in the `data/location_files` folder)

```python
loc_list = ['Sahara','Oregon','NewGuinea','Amazon','Siberia']
```

```python
ds_dict = {}

for loc in loc_list:
    print(loc)
    
    ds_temp = xr.open_dataset('data/location_files/'+loc+'_surface_energy_fluxes_CESM2_1pctCO2_0001-0150.nc')

        
    ds_dict[loc] = ds_temp.copy()
    
    del ds_temp
```

### Housekeeping: make a dictionary with helpful names and colors that we want to use in the bar graphs we're going to make of surface energy fluxes:

```python

attribute_dict = {}
attribute_dict['color'] = {}
attribute_dict['name'] = {}


attribute_dict['color']['rsds']='gold'
attribute_dict['color']['rsus']='gold'
attribute_dict['color']['rsns']='gold'
attribute_dict['color']['rlds']='purple'
attribute_dict['color']['rlus']='purple'
attribute_dict['color']['hfls']='blue'
attribute_dict['color']['hfss']='red'
attribute_dict['color']['ts']='black'


attribute_dict['name']['rsds']='Incident Shortwave Radiation'
attribute_dict['name']['rsus']='Reflected Shortwave Radiation'
attribute_dict['name']['rsns']='Absorbed Shortwave Radiation'
attribute_dict['name']['rlds']='Downwards Longwave Radiation'
attribute_dict['name']['rlus']='Upwards Longwave Radiation'
attribute_dict['name']['hfls']='Latent Heat'
attribute_dict['name']['hfss']='Sensible Heat'
attribute_dict['name']['ts']='Surface Temperature'
```

# Plot bar graphs of the breakdown of the terms of the surface energy budget for each location

We'll look at the average of the first 10 years of the simulation here. The simulation has CO$_2$ increasing at 1% per year, so it isn't _quite_ in equilibrium during the first 10 years, as the whole system is warming up due to increasing CO$_2$. 

Later in this notebook, you can see how surface fluxes are chaning in time (and thus with increasing atmospheric CO$_2$). 



### Plot the breakdown of the surface energy budget

Here, we plot the "energy in" terms of the surface energy budget on the left (incoming shortwave radiation and incoming longwave radiation). On the left, we plot the "energy out" terms of the surface energy budget (reflected shortwave, emitted longwave, sensible heat, and latent heat). 


The variable names of the surface energy budget follow the naming convention from the Coupled Model Intercomparison Project, and are as follows:   
- *rsds*: Radiation Shortwave Down Surface   (downwards shortwave at the surface, positive downwards/into of the surface)
- *rsus*: Radiation Shortwave Up Surface     (reflected shortwave at the surface, positive upwards/out of the surface)
- *rlds*: Radiation Longwave Down Surface    (downwards longwave at the surface, positive downwards/into of the surface)
- *rlus*: Radiation Longwave Up Surface      (emitted longwave from the surface, positive upwards/out of the surface)
- *hfss*: Heat Flux Sensible Surface         (sensible heat flux, positive upwards/out of the surface)
- *hfls*: Heat Flux Latent Surface           (latent heat flux, positive upwards/out of the surface)



```python

for loc in ['Sahara','Oregon','Amazon','NewGuinea','Siberia']:

    ds = ds_dict[loc].sel(year=slice(1,10)).mean('year')

    fig = plt.figure(figsize=[10,6])
    
    # energy into the land surface:
    SW_down = float(ds['rsds'].values)
    SW_up = float(ds['rsus'].values)
    net_SW = float((ds['rsds'] - ds['rsus']).values)
    LW_down = float(ds['rlds'].values)
    
    # energy out of the land surface: 
    LW_up = float(ds['rlus'].values)
    LH = float(ds['hfls'].values)
    SH = float(ds['hfss'].values)
    
    labels = ['SW$_{down}$',  'LW$_{down}$', 'SW$_{up}$','LW$_{up}$', 'SH', 'LH']
    values = [SW_down,LW_down,SW_up,LW_up,SH,LH]
    colors = [attribute_dict['color']['rsds'],attribute_dict['color']['rlds'],
              attribute_dict['color']['rsus'],attribute_dict['color']['rlus'],
              attribute_dict['color']['hfss'],attribute_dict['color']['hfls'] ]
    
    y_pos = np.arange(len(values))

    # Create bars
    plt.bar(y_pos, values,color=colors,edgecolor = 'k')

    # Create names on the x-axis
    plt.xticks(y_pos, labels)
    
    
    # add text showing the exact value of each bar
    ax = plt.gca()
    for i, v in enumerate(values):
        ax.text(i - 0.2 , v + 10, '%1.0f'%v, color='k', fontweight='bold',fontsize=14)
    
    
    # Label Energy In / Energy Out sides of graph and add a dividing line
    plt.ylim(0,540)
    ylim = ax.get_ylim()
    plt.plot([(y_pos[2] + y_pos[1])/2 , (y_pos[2] + y_pos[1])/2 ],
             [ylim[0], ylim[1]],
             'k--')
    plt.grid(':',zorder=-5,axis='y')
    
    plt.text(1/5,0.92,'Energy In',ha='center',transform=ax.transAxes,fontsize=18,fontweight='bold')
    plt.text(3.5/5,0.92,'Energy Out',ha='center',transform=ax.transAxes,fontsize=18,fontweight='bold')
    
    ax.tick_params(axis='both',labelsize=16)
    
    
    # make a string of the latitude/longitude to add to the figure:
    if float(ds.lon.values)>180:
        lon_str = '%1.1f deg W'%(360 - float(ds.lon.values))
    else:
        lon_str = '%1.1f deg E'%float(ds.lon.values)
        
    if float(ds.lat.values)>=0:
        lat_str = '%1.1f deg N'%(float(ds.lat.values))
    else:
        lat_str = '%1.1f deg S'%np.abs(float(ds.lat.values))
    
    # Label the axes and give figure a title
    plt.title('Average Surface Energy Fluxes,\n%s: %s, %s'%(loc,lon_str,lat_str),fontsize=18,y=1.02)
    
    plt.ylabel('W/m$^2$',fontsize=16)
    plt.xlabel('Surface Energy Budget Term',fontsize=16)
    
   
    
    plt.show()
    
    plt.close()
    
    
    
```

### Plot the total Energy In and Energy Out of the land surface 

```python

for loc in ['Sahara','Oregon','Amazon','NewGuinea','Siberia']:

    ds = ds_dict[loc].sel(year=slice(1,10)).mean('year')

    fig = plt.figure(figsize=[10,6])
    
    Ein = float( ds['rsds'].values + ds['rlds'].values )
    Eout = float( ds['rsus'].values + ds['rlus'].values + ds['hfls'].values + ds['hfss'].values )
    
    
    labels = ['E$_{in}$',  'E$_{out}$',]
    values = [Ein,Eout,]
    colors = ['black','darkgray',]
    
    y_pos = np.arange(len(values))

    # Create bars
    plt.bar(y_pos, values,color=colors,edgecolor = 'k')

    # Create names on the x-axis
    plt.xticks(y_pos, labels)
    
    
    # add text showing the exact value of each bar
    ax = plt.gca()
    for i, v in enumerate(values):
        ax.text(i - 0.1 , v + 10, '%1.0f'%v, color='k', fontweight='bold',fontsize=14)
    
    
    # Label Energy In / Energy Out sides of graph and add a dividing line
    plt.ylim(0,800)
    ylim = ax.get_ylim()
    plt.plot([(y_pos[1] + y_pos[0])/2 , (y_pos[1] + y_pos[0])/2 ],
             [ylim[0], ylim[1]],
             'k--')
    plt.grid(':',zorder=-5,axis='y')
    
    plt.text(1/4,0.92,'Energy In',ha='center',transform=ax.transAxes,fontsize=18,fontweight='bold')
    plt.text(3/4,0.92,'Energy Out',ha='center',transform=ax.transAxes,fontsize=18,fontweight='bold')
    
    ax.tick_params(axis='both',labelsize=16)
    
    
    # make a string of the latitude/longitude to add to the figure:
    if float(ds.lon.values)>180:
        lon_str = '%1.1f deg W'%(360 - float(ds.lon.values))
    else:
        lon_str = '%1.1f deg E'%float(ds.lon.values)
        
    if float(ds.lat.values)>=0:
        lat_str = '%1.1f deg N'%(float(ds.lat.values))
    else:
        lat_str = '%1.1f deg S'%np.abs(float(ds.lat.values))
    
    # Label the axes and give figure a title
    plt.title('Average Surface Energy Fluxes,\n%s: %s, %s'%(loc,lon_str,lat_str),fontsize=18,y=1.02)
    
    plt.ylabel('W/m$^2$',fontsize=16)
    plt.xlabel('Surface Energy Budget Term',fontsize=16)
    
   
    
    plt.show()
    
    plt.close()
    
    
    
```

## From here, we can look at how the trends in the surface tempeartures and the surface energy budget terms are changing with increased atmospheric CO$_2$

```python


for loc in ['Sahara','Oregon','Amazon','NewGuinea','Siberia']:

    ds = ds_dict[loc]

    years = ds.year.values


    fig, axes = plt.subplots(1,1,figsize=(6,4))

    for var in ['ts',]:

        if var in ['rsds','rsus','rsns','rlds','rlus','hfls','hfss']:
            units = 'W/m$^2$'
        elif var in ['ts']:
            units = 'K'
        else:
            units = ds_dict[loc][var].units

        ax = axes

        slope, intercept, r, p, se = stats.linregress(ds.year,ds[var])

        ax.plot(ds.year.values,ds[var],color=attribute_dict['color'][var],label=var)
        ax.plot(ds.year.values,intercept + ds.year.values*slope,'--',color=attribute_dict['color'][var],label='slope = %1.2f %s/year;\np = %1.3f'%(slope,units,p))

        ax.legend(fontsize=12)
        ax.grid()
        

        plt.xlabel('year',fontsize=14)
        plt.ylabel(units,fontsize=14)


    fig.suptitle('Surface Temperature (%s)'%loc,fontsize=16)

    plt.show()
    plt.close()
```

```python
run = '1pctCO2'

for loc in ['Sahara','Oregon','Amazon','NewGuinea','Siberia']:
    

    ds = ds_dict[loc]

    years = ds.year.values


    # fig = plt.subplots(figsize=(16,8))
    fig, _axes = fig, axes = plt.subplots(1,1,figsize=(6,4))
#     axes = _axes.flatten()

#     for var in ['Ein','Eout','Eres']:

    units = 'W/m$^2$'

    ax = _axes


    energy_in = ds['rsns'] + ds['rlds']

    energy_out = ds['rlus'] + ds['hfls'] + ds['hfss']

#     residual = ds['rsns'] + ds['rlds'] - ( ds['rlus'] + ds['hfls'] + ds['hfss'] )

    
    data = energy_in
    slope, intercept, r, p, se = stats.linregress(ds.year,data)
    ax.plot(ds.year.values,data,color='k',label='Energy In')
    ax.plot(ds.year.values,intercept + ds.year.values*slope,'--',color='k',label='slope = %1.2f %s/year;\np = %1.3f'%(slope,units,p))

    data = energy_out
    slope, intercept, r, p, se = stats.linregress(ds.year,data)
    ax.plot(ds.year.values,data,color='gray',label='Energy Out',alpha=0.6,zorder=0)
    ax.plot(ds.year.values,intercept + ds.year.values*slope,'--',color='gray',alpha=0.6,label='slope = %1.2f %s/year;\np = %1.3f'%(slope,units,p),zorder=0)



#     data = residual
#     slope, intercept, r, p, se = stats.linregress(ds.year,data)
#     ax.plot(ds.year.values,data,':',color='gray',label='Residual',alpha=0.5)
#     ax.plot(ds.year.values,intercept + ds.year.values*slope,'--',color='k',label='slope = %1.2f %s/year;\np = %1.3f'%(slope,units,p))


    ax.legend(fontsize=12,loc='upper left',bbox_to_anchor=[1.,1.])
    ax.grid()

    # ax.set_title(var,fontsize=14)

    plt.xlabel('year',fontsize=14)
    plt.ylabel(units,fontsize=14)


    fig.suptitle('Extra Energy Passing Through the Surface (%s)'%loc,fontsize=16)

    plt.show()
    plt.close()

```

### Note: 

if you run the above cell for Siberia, you'll see that the black (Energy In) and gray (Energy Out) lines don't overlap quite as much as they do for the other locations. This shows that the land surface in the high latitudes isn't quite in equilibrium as the climate warms - rather, the land is taking up energy. Permafrost is warming so there is more energy going into the surface than there is leaving the surface. 


```python
run = '1pctCO2'

for loc in ['Sahara','Oregon','Amazon','NewGuinea','Siberia']:

    ds = ds_dict[loc]

    years = ds.year.values


    # fig = plt.subplots(figsize=(16,8))
    fig, _axes = plt.subplots(2,3,figsize=(16,8))
    axes = _axes.flatten()

    i = 0
    for var in ['rsds','rsus','hfss','rlds','rlus','hfls',]:

        if var in ['rsds','rsus','rsns','rlds','rlus','hfls','hfss']:
            units = 'W/m$^2$'
        elif var in ['ts']:
            units = 'K'
        else:
            units = ds_dict[loc][var].units

        ax = axes[i]

        slope, intercept, r, p, se = stats.linregress(ds.year,ds[var])

        ax.plot(ds.year.values,ds[var],color=attribute_dict['color'][var],label=var)
        ax.plot(ds.year.values,intercept + ds.year.values*slope,'--',color=attribute_dict['color'][var],label='slope = %1.2f %s/year;\np = %1.3f'%(slope,units,p))

        ax.legend(fontsize=12)
        ax.grid()
        
        ax.set_title(attribute_dict['name'][var],fontsize=14)

        plt.xlabel('year',fontsize=14)
        plt.ylabel(units,fontsize=14)

        i = i+1

    fig.suptitle('Surface Energy Fluxes (%s)'%loc,fontsize=16)

    plt.show()
    plt.close()
```

## We can also plot the difference in the surface energy fluxes (like the bar graphs we started with) for the last 10 years of the run (high CO2) minus the first 10 years of the run (low CO2)

```python

for loc in ['Sahara','Oregon','Amazon','NewGuinea','Siberia']:

    ds = ds_dict[loc].sel(year=slice(141,150)).mean('year') - ds_dict[loc].sel(year=slice(1,10)).mean('year')

    fig = plt.figure(figsize=[10,6])
    
    # energy into the land surface:
    SW_down = float(ds['rsds'].values)
    SW_up = float(ds['rsus'].values)
    net_SW = float((ds['rsds'] - ds['rsus']).values)
    LW_down = float(ds['rlds'].values)
    
    # energy out of the land surface: 
    LW_up = float(ds['rlus'].values)
    LH = float(ds['hfls'].values)
    SH = float(ds['hfss'].values)
    
    labels = ['SW$_{down}$',  'LW$_{down}$', 'SW$_{up}$','LW$_{up}$', 'SH', 'LH']
    values = [SW_down,LW_down,SW_up,LW_up,SH,LH]
    colors = [attribute_dict['color']['rsds'],attribute_dict['color']['rlds'],
              attribute_dict['color']['rsus'],attribute_dict['color']['rlus'],
              attribute_dict['color']['hfss'],attribute_dict['color']['hfls'] ]
    
    y_pos = np.arange(len(values))

    # Create bars
    plt.bar(y_pos, values,color=colors,edgecolor = 'k')

    # Create names on the x-axis
    plt.xticks(y_pos, labels)
    
    
    # add text showing the exact value of each bar
    ax = plt.gca()
    for i, v in enumerate(values):
        if v>=0:
            ax.text(i - 0.2 , v + 2, '%1.0f'%v, color='k', fontweight='bold',fontsize=14)
        else:
            ax.text(i - 0.2 , v - 5, '%1.0f'%v, color='k', fontweight='bold',fontsize=14)
    
    
    # Label Energy In / Energy Out sides of graph and add a dividing line
    plt.ylim(-30,75)
    ylim = ax.get_ylim()
    plt.plot([(y_pos[2] + y_pos[1])/2 , (y_pos[2] + y_pos[1])/2 ],
             [ylim[0], ylim[1]],
             'k--')
    plt.grid(':',zorder=-5,axis='y')
    
    plt.text(1/5,0.92,'Energy In',ha='center',transform=ax.transAxes,fontsize=18,fontweight='bold')
    plt.text(3.5/5,0.92,'Energy Out',ha='center',transform=ax.transAxes,fontsize=18,fontweight='bold')
    
    ax.tick_params(axis='both',labelsize=16)
    
    
    # make a string of the latitude/longitude to add to the figure:
    if float(ds.lon.values)>180:
        lon_str = '%1.1f deg W'%(360 - float(ds.lon.values))
    else:
        lon_str = '%1.1f deg E'%float(ds.lon.values)
        
    if float(ds.lat.values)>=0:
        lat_str = '%1.1f deg N'%(float(ds.lat.values))
    else:
        lat_str = '%1.1f deg S'%np.abs(float(ds.lat.values))
    
    # Label the axes and give figure a title
    plt.title('Average $\Delta$ Surface Energy Fluxes,\nHigh CO2 (years 141-150) - Low CO2 (years 1-10)\n%s: %s, %s'%(loc,lon_str,lat_str),fontsize=18,y=1.02)
    
    plt.ylabel('W/m$^2$',fontsize=16)
    plt.xlabel('Surface Energy Budget Term',fontsize=16)
    
   
    
    plt.show()
    
    plt.close()
    
    
    
```
